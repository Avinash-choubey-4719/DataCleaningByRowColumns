# CHECKING FOF THE PERCENTAGE OF THE NULL VALUES

```python
null_value_perc_list = dataFrame.isnull().sum()/dataFrame.shape[0] * 100
```

```python
null_value_perc_list
```

# REMOVE THE COLUMN CONTAINING HIGH NUMBER OF NULL VALUES

```python
column_to_be_dropped = null_value_perc_list[null_value_perc_list >15].keys()
```

```python
column_to_be_dropped
```

```python
dataFrame_dropped_columns = dataFrame.drop(columns = column_to_be_dropped)
```

```python
dataFrame_dropped_columns.head()
```

```python
dataFrame_dropped_columns.isnull().sum()/dataFrame_dropped_columns.shape[0] * 100
```

```python
dataFrame_dropped_columns.shape
```

# NOW WE WILL REMOVE THE PARTICULAR ROW CONTAINING NULL VALUE

```python
dataFrame_dropped_rows = dataFrame_dropped_columns.dropna()
```

```python
dataFrame_dropped_rows.isnull().sum()/dataFrame_dropped_rows.shape[0] * 100
```

# VISUALIZING THE HEATMAP IN ALL THE CASES AS BELOW

```python
plt.figure(figsize = (25, 25))
sns.heatmap(dataFrame.isnull(), color = 'blue')
```

```python
plt.figure(figsize = (25, 25))
sns.heatmap(dataFrame_dropped_columns.isnull())
```

```python
plt.figure(figsize = (25, 25))
sns.heatmap(dataFrame_dropped_rows.isnull())
```

# STORING THE COLUMN NAMES HAVING THE PARTICULAR DATA TYPE

```python
list_int_columns = dataFrame_dropped_rows.select_dtypes(['int', 'float']).columns
```

```python
list_int_columns
```

```python
list_object_columns = dataFrame_dropped_rows.select_dtypes(['object']).columns
```

# NOW BELOW CODE WILL CHECK FOR THE CAUSE OF DATA CLEANING

```python
list_object_columns
```

```python
plt.figure(figsize = (25, 25))
sns.distplot(dataFrame['LotArea'])
sns.distplot(dataFrame_dropped_rows['LotArea'])
```

```python
def show_distplot_orig(name):
    return sns.distplot(dataFrame[name])
```

```python
def show_distplot_new(name):
    return sns.distplot(dataFrame_dropped_rows[name])
```

```python
for item in list_int_columns:
    plt.figure(figsize = (25, 25))
    show_distplot_orig(item)
    show_distplot_new(item)
```

```python
plt.figure(figsize = (25, 25))
for i, item in enumerate(list_int_columns):
    plt.subplot(9, 4, i + 1)
    sns.distplot(dataFrame[item], bins = 20)
    sns.distplot(dataFrame_dropped_rows[item], bins = 20)
```

```python
pd.concat([dataFrame['Street'].value_counts()/dataFrame.shape[0] * 100, 
          dataFrame_dropped_columns['Street'].value_counts()/dataFrame_dropped_columns.shape[0] * 100,
          dataFrame_dropped_rows['Street'].value_counts()/dataFrame_dropped_rows.shape[0] * 100], 
          axis = 1, 
          keys = ['original', 'new'])
```

```python
def show_difference(name):
    return pd.concat([dataFrame[name].value_counts()/dataFrame.shape[0] * 100, 
                     dataFrame_dropped_columns[name].value_counts()/dataFrame_dropped_columns.shape[0] * 100,
                     dataFrame_dropped_rows[name].value_counts()/dataFrame_dropped_rows.shape[0] * 100], 
                     axis = 1,
                     keys = [name + 'original', name + 'new'])
```

```python
for item in list_object_columns:
    show_difference(item)
```
